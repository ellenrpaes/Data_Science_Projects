### Homicídio Doloso

O homicídio doloso ocorre quando uma pessoa tira a vida da outra de forma intencional ou quando assume o risco de fazê-lo. Há duas formas de classificar esse crime:

*   Dolo direto: Quando o indivíduo deseja matar sua vítima
*   Dolo indireto: Quando o indivíduo náo tinha a intenção de matar mas organizou algum evento que causou a morte de alguém.

O homicídio doloso está previsto no artigo 121 do Código Penal Brasileiro.


```
## Tabela Resumo
mean_hom_doloso = df.hom_doloso.mean()
min_hom_doloso = df.hom_doloso.min()
max_hom_doloso = df.hom_doloso.max()

print('Média de Homicídios por mês: {:.1f}'.format(mean_hom_doloso))
print('Qual foi o menor resultado para um mês: {:.1f} homicidios dolosos'.format(min_hom_doloso))
print('QUal foi o maior resultado para um mês: {:.1f} homicidios dolosos'.format(max_hom_doloso))
```

É possível observar pela análise do histograma que:

*   Houveram quase 70 meses com 500 a 550 homicídios dolosos registrados ao longo da série estudada.
*   Houveram quase 20 meses com registros superiores a 650 homicídios mês.
*   Somente em pouco mais de 20 meses os homicídios ficaram entre 250 e 330 meses aproximadamente.



```
df.hom_doloso.hist();
```

É possível observar pela série histórica que esse tipo de crime está reduzindo ao longo dos anos. Se compararmos 2019 com 2018, temos uma redução de 19% no número de registros. O resultado com pior mês de homicídios dolosos aconteceu em Janeiro/1995 com 831 casos, conforme abaixo. Já o melhor mês ocorreu em Junho/2015 com 272 casos.
